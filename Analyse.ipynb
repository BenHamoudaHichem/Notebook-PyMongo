```python
#Pour travailler mieux avec les framework aggregate on va créer une autre collection meteoriteNASA
#Nous avons importer des données concernant "meteorite-fall" du NASA
#NB j'ai utiliser aggregate dans l'entrée 11 pour transferer les données vers une autre collection
from urllib.request import urlopen

import json

# réccupération d'une api data et stockage
response = urlopen("https://data.nasa.gov/resource/y77d-th95.json")

data_json = json.loads(response.read())
c_meteoriteNASA=pointeur.meteoriteNASA
c_meteoriteNASA.insert_many(data_json)
```

```python
#Data Cleannig

for item in c_meteoriteNASA.find({}):
    try:  
        item['year']=dateutil.parser.parse(item['year'])
        item['reclat']=float(item['reclat'])
        item['reclong']=float(item['reclong'].replace('U00002013', '-'))
    except KeyError:    
        print('item non-reconnue')
    c_meteoriteNASA.replace_one({'_id':item['_id']},item)
for item in c_meteoriteNASA.find({"mass": {"$exists": True}}):
    try:
        item['mass']=float(item['mass'])
    except KeyError:
        print('item non-reconnue')
    c_meteoriteNASA.replace_one({"_id":item['_id']},item)

# Affichage du collection

for item in c_meteoriteNASA.find({}):
    pprint.pprint(item)
```

```python
#Afin d'obtenir des résultats notables, j'ai décidé de modifier l'attribut year
#Comme vous remarquer le mois et le le jour ont toujous la valeur 1
#donc nous allon regénerer seulement le jour et le mois
finalResult= c_meteoriteNASA.aggregate(
   [
     {
      "$project":
         {
           "year": { "$year": "$year" },
           "month": { "$month": "$year" },
           "day": { "$dayOfMonth": "$year" },
           
         }
     }
   ]
)
from random import randrange
from datetime import timedelta

def random_date(start, end):
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    random_second = randrange(int_delta)
    return start + timedelta(seconds=random_second)

from datetime import datetime
for item in finalResult:
    if item['year'] is None or item['year'] < 1000:
        print(c_meteoriteNASA.delete_one({"_id":item["_id"]}).deleted_count,"Enregistrement supprimé ")
    else:
        
        d1 = datetime.strptime('1/1/'+str(item['year'])+' 1:30 PM', '%d/%m/%Y %I:%M %p')
        d2 = datetime.strptime('31/12/'+str(item['year'])+' 4:50 PM', '%d/%m/%Y %I:%M %p')
        c_meteoriteNASA.update_one({"_id":item['_id']},{"$set":{"year":random_date(d1, d2)}})

```

```python
#Voici le resultat
finalResult= c_meteoriteNASA.aggregate(
   [
     {
      "$project":
         {
           "year": { "$year": "$year" },
           "month": { "$month": "$year" },
           "day": { "$dayOfMonth": "$year" },
           
         }
     }
   ]
)
for i in finalResult:
    print(i)
```

```python
#trouver le nombre de meteorie par class
finalResult=c_meteoriteNASA.aggregate( [
   { "$group": { "_id": "$recclass", "Total": { "$sum": 1 } } }
] )
for item in finalResult:
    print(item)
```

```python
#Trier par la masse
finalResult=c_meteoriteNASA.aggregate( [
   { "$group": { "_id": {"year":{"$year":"$year"}}, "totalMasse": { "$sum": "$mass" } } },{"$sort":{"totalMasse":-1}},
] )
for item in finalResult:
    print(item)
```

```python
#La résultat précedente retourne la mass None si elle n'existe pas dans un document
#on va ajouter $match
finalResult=c_meteoriteNASA.aggregate( [
    {"$match":{"mass":{"$exists":True}}},
    { "$group": { "_id": {"year":{"$year":"$year"}}, "maxMasse": {"$max": "$mass" } } },{"$sort":{"maxMasse":-1}},])

for item in finalResult:
    print(item)
```
